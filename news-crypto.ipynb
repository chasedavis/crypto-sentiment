```python
# Fix typing
clean_responses = []
for response in responses:
    keys = response.keys()
    new_response = {}
    for key in keys:
        new_response[type_fix(key)] = type_fix(response[key])
    new_response['publishedAt'] = parse(new_response['publishedAt'])
    clean_responses.append(new_response )
```

```python
print "Number of headlines: ", len(responses), "\n\n"

# Write data to csv (appending)
with open("news-ethereum6.csv", "a") as f:
    w = csv.DictWriter(f, clean_responses[0].keys())
    w.writeheader()
    w.writerows(clean_responses)
```
